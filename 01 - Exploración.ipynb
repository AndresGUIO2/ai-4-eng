## Análisis inicial de datos

```python
# Cargar el conjunto de datos
df = pd.read_csv('udea-ai-4-eng-20251-pruebas-saber-pro-colombia/train.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Cantidad de registros
print(f"Cantidad de registros: {df.shape[0]}")

# Cantidad de columnas
print(f"Cantidad de columnas: {df.shape[1]}")

# Cantidad de valores nulos
print(f"Cantidad de valores nulos: {df.isnull().sum().sum()}")
```

```python
# Lista de estudiantes por programa académico
estudiantes_por_programa = df['ESTU_PRGM_ACADEMICO'].value_counts()
print(estudiantes_por_programa)
```

```python
# Lista de estudiantes por estrato
estudiantes_por_estrato = df['FAMI_ESTRATOVIVIENDA'].value_counts()
print(estudiantes_por_estrato)
```

```python
#Programas académicos más populares
programas_mas_populares = df['ESTU_PRGM_ACADEMICO'].value_counts().head(10)
print(programas_mas_populares)
```
